```python
import requests
import json
import pandas as pd

r = requests.get("https://gamma-api.polymarket.com/events?slug=what-will-trump-say-during-inauguration-speech")
response = r.json()

response
```

```python
# Server: AWS US-East-2
```

```python
market_data = [] 

# Iterate through markets and append data
for market in response[0]['markets']:
   market_info = {
       'question': market['question'],
       'market_id': market['id'],
       'token_id_YES': json.loads(market['clobTokenIds'])[0],
       'token_id_NO': json.loads(market['clobTokenIds'])[1]
   }
   market_data.append(market_info) 

df = pd.DataFrame(market_data)
df 
```

```python
df.to_csv('trump_inauguration_markets.csv', index=False)
print("Data exported to trump_inauguration_markets.csv")
```
